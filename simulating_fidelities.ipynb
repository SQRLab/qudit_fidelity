```python
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
connected_distance = 1 
trapped_distance_1 = 1.3 
trapped_distance_2 = 1.6
superconducting_distance = 1.7
```

```python
def compute_fidelity_lower_bounds(deltas, distance):
    fidelities = []
    for delta in deltas:
        fidelities.append(1 / distance ** 2 * delta ** 2)
    return fidelities
```

```python
# plotting code 
deltas = np.linspace(0, 1, 100)
fidelities_connected = compute_fidelity_lower_bounds(deltas, connected_distance)
fidelities_trapped_1 = compute_fidelity_lower_bounds(deltas, trapped_distance_1)
fidelities_trapped_2 = compute_fidelity_lower_bounds(deltas, trapped_distance_2)
fidelities_superconducting = compute_fidelity_lower_bounds(deltas, superconducting_distance)
```

```python
sns.set_theme(style="darkgrid", font='montserrat')
```

```python
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]
labels = ["connected", "potential_trap", "current_trap", "superconducting"]
fidelities = [fidelities_connected, fidelities_trapped_1, fidelities_trapped_2, fidelities_superconducting]
sns.set_palette(sns.color_palette(colors))
fig, ax = plt.subplots()
for i, fidelity in enumerate(fidelities):
    ax.plot(deltas, fidelity, label=labels[i])
ax.set(xlabel="Delta", ylabel="Fidelity", title="Fidelity lower bounds")
ax.legend()
plt.show()
```
